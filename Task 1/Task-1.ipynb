```python
import numpy as np
```

# CREATING TWO RANDOM 2x2 MATRICES

```python
np.random.seed(3)
```

```python
A = np.random.randint(low=0, high=11, size=(2,2))
B = np.random.randint(low=2, high=13, size=(2,2))

print(f"MATRIX A is:\n{A}")
print(f"Shape of MATRIX A is {A.shape}")
print()
print(f"MATRIX B is:\n{B}")
print(f"Shape of MATRIX B is {B.shape}")
```

## Element Wise Multiplication

```python
ew_mul = A*B
print(ew_mul)
print(f"Shape of the matrix is: {ew_mul.shape}")
```

## Matrix Multiplication

```python
mat_mul = A@B
print(mat_mul)
print(f"Shape of the matrix is: {mat_mul.shape}")
```

# COMPUTING INVERSE AND DETERMINANT

## Determinant Calculations

```python
det_A = np.linalg.det(A)
print(det_A)
```

```python
det_B = np.linalg.det(B)
print(det_B)
```

## Inverse Calculations

```python
inv_A = np.linalg.inv(A)
print(inv_A)
print(f"Shape of the matrix is: {inv_A.shape}")
```

```python
inv_B = np.linalg.inv(B)
print(inv_B)
print(f"Shape of the matrix is: {inv_B.shape}")
```

# NORMALIZING A 1D ARRAY

```python
arr_one_D = np.array([10,14,37,50])
print(f"Shape of the matrix is: {arr_one_D.shape}")
```

```python
normalized_arr = (arr_one_D - arr_one_D.min()) / (arr_one_D.max() - arr_one_D.min())
print(normalized_arr)
print(f"Shape of the matrix is: {normalized_arr.shape}")
```

# DEMONSTRATING BROADCASTING

```python
column_mean = np.mean(A, axis=0)
print(f"Shape of the column mean is: {column_mean.shape}")

new_A = A - column_mean

print()
print(new_A)
print(f"Shape of the matrix is: {new_A.shape}")
```
