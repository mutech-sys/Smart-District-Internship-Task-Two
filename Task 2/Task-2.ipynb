```python
import pandas as pd
```

# LOADING DATASET

```python
df = pd.read_csv("sales_practice.csv")
```

# INSEPECTING DATASET

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

# ADDING 'revenue' COLUMN

```python
df['revenue'] = df['price'] * df['quantity']
```

# FILTERING DATASET

```python
north_df = df[df['region'] == 'North']
```

```python
north_df.head()
```

# GROUPING DATA BASED ON PRODUCTS

```python
grp = df.groupby(['product'])
```

## Total Revenue

```python
product_revenue = grp['revenue'].sum().reset_index()
```

# TOP 5 PRODUCTS

```python
top_5 = product_revenue.sort_values(by='revenue', ascending=False)
```

```python
top_5
```

# MERGRING DATASETS

```python
final_result = pd.merge(top_5, df, on="product", how="left")
```

# CREATING A CSV

```python
final_result.to_csv("north_products.csv", index=False)
```
