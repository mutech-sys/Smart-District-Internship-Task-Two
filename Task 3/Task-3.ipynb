```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

sns.set_theme()
```

# LOADING DATASET

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

# LOOKING AT DATASET

```python
df.head()
```

```python
df.info()
```

# PLOTTING HISTOGRAMS FOR ALL NUMERIC COLUMNS USING SEABORN

## Histogram for Age column

```python
sns.histplot(df['Age'], kde=True, bins=[i for i in range(0, 85, 2)])
plt.savefig("./Plots/Histogram_Age")
```

## Histogram for SibSp (Sibling or Spouse) column

```python
sns.histplot(df['SibSp'], kde=True)
plt.savefig("./Plots/Histogram_SibSp")
```

## Histogram for Parch (Parent or Child) column

```python
sns.histplot(df['Parch'], kde=True)
plt.savefig("./Plots/Histogram_Parch")
```

## Histogram for Fare column

```python
sns.histplot(df['Fare'], kde=True)
plt.savefig("./Plots/Histogram_Fare")
```

# PLOTTING BOX PLOT FOR ALL NUMERICAL COLUMNS

## Box plot for Age column

```python
sns.boxplot(x=df['Age'])
plt.savefig("./Plots/BoxPlot_Age")
```

## Box plot for SibSp (Sibiling or Spouse) column

```python
sns.boxplot(x=df['SibSp'])
plt.savefig("./Plots/BoxPlot_SibSp")
```

## Box plot for Parch (Parent or Child) column

```python
sns.boxplot(x=df['Parch'])
plt.savefig("./Plots/BoxPlot_Parch")
```

## Box plot for Fare column

```python
sns.boxplot(x=df['Fare'])
plt.savefig("./Plots/BoxPlot_Fare")
```

# PLOTTING HEATMAP

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.savefig("./Plots/Heatmap")
```

# PLOTTING PAIR PLOT

```python
sns.pairplot(df[['Age', 'SibSp', 'Parch', 'Fare']])
plt.savefig("./Plots/PairPlot")
```

# PLOTTING COUNT PLOT

## Count plot for Survived column

```python
sns.countplot(x='Survived', data=df)
plt.savefig("./Plots/CountPlot_Survived")
```

## Count plot for Pclass (Passenger class) column

```python
sns.countplot(x='Pclass', data=df)
plt.savefig("./Plots/CountPlot_Pclass")
```

## Count plot for Sex column

```python
sns.countplot(x='Sex', data=df)
plt.savefig("./Plots/CountPlot_Sex")
```

## Count plot for Embarked column

```python
sns.countplot(x='Embarked', data=df)
plt.savefig("./Plots/CountPlot_Embarked")
```
