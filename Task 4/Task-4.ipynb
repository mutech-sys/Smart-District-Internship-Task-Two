# DISTRIBUTION OF TARGET COLUMN 'Survived'

```python
sns.histplot(df['Survived'])
```

# FINDING CORRELATIONS USING HEATMAP

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
```

The most correlated column with the target column is 'Pclass'.

# FINDING OUTLIERS USING BOX PLOT

## Box plot for Age column

```python
sns.boxplot(x='Age', data=df)
```

There are some outliers in the column Age

## Box plot for Fare column

```python
sns.boxplot(x='Fare', data=df)
```

There are many outliers in the Fare column

# FINDING CLASS IMBALANCES

## Count plot for Sex column

```python
sns.countplot(x='Sex', data=df)
```

```python
df['Sex'].value_counts()
```

There is a class imbalance. Because of more male than female.

## Count plot for Pclass

```python
sns.countplot(x='Pclass', data=df)
```

```python
df['Pclass'].value_counts()
```

There is a class imbalance. Because there are more members of 3 Pclass then other two.

## Count plot for SibSp column

```python
sns.countplot(x='SibSp', data=df)
```

```python
df['SibSp'].value_counts()
```

There is a big class imbalance. Because there are significantly more values with the class 0 in SibSp

## Count plot for Parch column

```python
sns.countplot(x='Parch', data=df)
```

```python
df['Parch'].value_counts()
```

There is a similar class imbalance as the SibSp column

## Count plot for Embarked column

```python
sns.countplot(x='Embarked', data=df)
```

```python
df['Embarked'].value_counts()
```

There is a class imbalance. Because the class **S** is significantly more than the other two classes
