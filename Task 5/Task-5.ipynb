# DATA CLEANING

## Converting 'Quantity' Column to Numerical form

```python
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
```

## Converting 'Price Per Unit' Column to Numerical form

```python
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
```

## Converting 'Total Spent' Column to Numerical form

```python
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
```

## Converting 'Transaction Date' Column to Numerical form

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
```

## Replacing all the 'ERROR' and 'UNKNOWN' with `pd.nan`

```python
df.replace(['ERROR', 'UNKNOWN'], pd.NA, inplace=True)
```
