```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
import joblib
```

# LOADING DATA SET

```python
df = pd.read_csv('Churn.csv')
```

```python
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['Contract'].value_counts()
```

```python
X = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
num_cols = [
    'tenure', 
    'MonthlyCharges', 
    'TotalCharges' 
]

ord_cols = [
    'Contract'
]

nom_cols = [
    'gender', 
    'SeniorCitizen', 
    'Partner', 
    'Dependents', 
    'PhoneService', 
    'MultipleLines', 
    'InternetService', 
    'OnlineSecurity', 
    'OnlineBackup', 
    'DeviceProtection', 
    'TechSupport', 
    'StreamingTV', 
    'StreamingMovies', 
    'PaperlessBilling', 
    'PaymentMethod'
]
```

```python
contract_cats = [['Month-to-month', 'One year', 'Two year']]
```

```python
num_transformer = StandardScaler()
```

```python
nom_transformer = OneHotEncoder(drop='first', handle_unknown='ignore')
```

```python
ord_transformer = OrdinalEncoder(categories=contract_cats, 
                                 handle_unknown='use_encoded_value', 
                                 unknown_value=-1)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_cols),
        ('nom', nom_transformer, nom_cols),
        ('ord', ord_transformer, ord_cols)
    ],
    remainder='drop' # Drops any columns not explicitly listed above (like 'Churn')
)
```

```python
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)
])
```

```python
X_processed = pipeline.fit_transform(df)
```

```python
joblib.dump(pipeline, 'churn_preprocessor.pkl')
print("✅ Pipeline successfully saved to 'churn_preprocessor.pkl'")
```
